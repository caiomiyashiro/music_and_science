# From Time to Frequency Domain - The Fourier Transform

In western music, when we play a chord in our instrument, we are playing a set of specific notes. For instance we can differentiate a C minor from a C major by it's third (E or Eb). But, what if we could decompose any kind of sound in its respective notes (frequencies)? That's basically the Fourier transform. Despite its fancy name, this transformation basically just answer you the following question: "`What are the frequencies that, together, compose the sound I'm seeing?`"

<img src="fourier.gif" style="height:100">  

<center>http://mriquestions.com/fourier-transform-ft.html</center>

Ok, when I said "*just answer x*" I was trying to make things simpler as they seemed. Let's take two examples of what the transformation can show us, lets hear two sounds, a whistle and a piano note, and lets see their Fourier transformation. If you don't want to understand the code below, it just takes the sound signal, perform the Fourier transform and plot the result, showing which frequencies (x-axis) are present in the signal:

```python
# https://stackoverflow.com/questions/4364823/how-do-i-obtain-the-frequencies-of-each-value-in-an-fft/4371627#4371627
# https://www.mathworks.com/help/matlab/ref/fft.html
# https://www.physicsforums.com/threads/what-is-fft-frequency-resolution.650931/

def plot_spectra(signal, Fs, max_freq=None):
    if(len(signal) % 2 != 0):
        signal = signal[:-1]
        
    n_half = int(len(signal)/2)                            
       
    transform = np.abs(np.fft.fft(signal)[:n_half])        # perform transformation and get first half points
    transform = transform/int(len(signal)/2)               # normalise

    # x-axis calculation
    freqs= Fs*np.arange(len(signal))/len(signal)
    freqs = freqs[:n_half]                                 # first half of the frequencies
    
    plt.plot(freqs, transform)                  
    plt.ylabel('Amplitude');
    plt.xlabel('Frequency (Hz)');
    if max_freq is None:
        max_x = Fs/2
    else:
        max_x = max_freq
    plt.xlim([0, max_x]);
```

# Short Time Fourier Transform - STFT 
  
Even though I tried to make the Fourier transform is easy, it actually has a lot of heavy mathematics envolved, just check its [Wikipedia Page](https://en.wikipedia.org/wiki/Fourier_transform). I'll try to keep most of the mathematics hidden but there is one thing that I can't skip for our melody detection system, which is the `Frequency Resolution`.  
  
If you look at my function to plot the signal spectra, you will see I do some maths when calculating what it's going to be show at the x-axis, well that's because what is on the x-axis depends on a few things besides the signal frequencies, more specifically, the signal size and it's sample frequency (how many sound samples per second were recorded in this sound). Given the signal of lengh L and sample frequency Fs, each index of the x-axis will represent the presence of a specific frequency of value $f = Fs/L$. For instance, if I have $L$ = 100 and $Fs$ = 200, each x-axis index will be every 2Hz. That means that every frequency in between 1Hz and 2Hz will fall inside of these 2 bins. Very nice, **but how this is important?**  
  
## Frequency Resolution and Window Size

If we want to detect frequency changes in a smaller scale than the frequency resolution, we won't be able to detect it. We saw how the frequency resolution depend on the sample frequency $Fs$ and the signal size $L$. When we are defining what's going to be the signal size we'll get for our STFT, we have to keep in mind what is the smallest frequency variation we want to be able to detect and adapt the signal size section accordingly (as we can't change $Fs$). This comes to an important concept in STFT, the **time-frequency trade-off**:
  
### Time-Frequency Trade-off  
  
In short terms, the smaller the window (signal slice), the better time resolution we have, as we will be able to do STFT in a lot more window sections but, as $L$ becomes small, we lose the frequency resolution. In our melody detection, we want to be able to identify at least 1 semitone of difference. If take the A4, at ~440 Hz as an example, by the [Equal Temperament](https://en.wikipedia.org/wiki/Equal_temperament), each semitone is dependent by the previous tone on a proportion of $\sqrt[12]2$. For instance, in order to perceive a half step from the A4 note, A#4 is going to be equal to $440 * \sqrt[12]2$ ~= 466 Hz.  
  
In human whistlening, the lowest note a person can whistle is around 700 Hz. So if we want to be able to detect the smallest change up, *i.e.*  a half step, we must be able to detect $700 * (\sqrt[12]2 - 1)$ ~= 42 Hertz. 

* Our sample frequency = 44100 = Fs  
  
* Our signal length = 19662 = len_signal

* Frequency resolution = len_signal/Fs = 2.24 Hz
  
If we want to obtain the smallest time window (better time resolution) but still be able to detect a variation of at least ~42 Hz, we just have to rearrange the frequency resolution equation:  

$$Fs/L = Fr ===> L = Fs/Fr $$
  
So that we want:

$$L = 44100/42 = 1.050 \space points$$

And if we want to convert the points to time domain, we just divide it by the sampling frequency:  

$$1.050 / Fs = 1.050/44100 ~= 0,024 \space seconds ~= 24 \space milliseconds$$   
  
**Ps**: This is, again, a simpler approach. If you want to read more on how we can better define the frequency resolution of the STFT, you'll have to read more on [Windowing Operations](http://support.ircam.fr/docs/AudioSculpt/3.0/co/Window%20Size.html#ecN1af)

Below is the STFT function that, given the audio signal and the sampling frequency, calculates the short time Fourier transform. For each time segment extracted from the signal, the traditional FT is applied and returned. By applying the FT in all window segments, the Scipy function returns a series of arrays:  

* **f** - 1d array - frequencies associated in the x-axis, as we showed before
* **t** - 1d array - time approximation of each time slice
* **Zxx** - matrix - frequencies for each time slice  
  
In order to facilitate the interpretation, we can plot the Spectrogram of the transformation. It show us the frequencies present during each time slice. Together, all the slices produce a really good approximation of frequency on time information:  

```python
# https://kevinsprojects.wordpress.com/2014/12/13/short-time-fourier-transform-using-python-and-numpy/
def stft_music(signal, Fs, print_fft=False):
    window_length_ms = 24
    window_length_points = int(window_length_ms*Fs/1000) # 3 rule

    f, t, Zxx = stft(signal, Fs, nperseg=window_length_points)

    MIN_HZ = 700
    MAX_HZ = 3000
    hz_filter = (f > MIN_HZ) & (f < MAX_HZ)
    f = f[hz_filter]
    Zxx = np.abs(Zxx)[hz_filter]
    
    t_total = len(signal)/Fs
    t = np.linspace(0, t_total, len(Zxx[0]))

    if(print_fft == True):
        plt.pcolormesh(t, f, Zxx, vmin=0, vmax=np.max(Zxx))
        plt.title('STFT Magnitude')
        plt.ylabel('Frequency [Hz]')
        plt.xlabel('Time [sec]')
        plt.show();
        
    return (f, t, Zxx)
    
s_3_norm = np.array(s_3) / max(s_3) # normalize audio amplitude to 1    
f1, t1, Zxx1 = stft_music(s_3_norm, Fs_3, True)    
```

The problem it seems is that extracting the maximum energy frequency allow the entry of too many artifacts, *i.e.*, noise. Even in some silent moments we could have inserted some random frequency just by pure noise from the microphone recording.  
  
In order to solve this, we can search only for "Whistled" frequencies. We know that a whistle produces an unique frequency (no harmonics) so most of the signal energy will be concentrated on this specific frequency. The function below check the peak energy frequency and its neighboors, defined as `peak_width` in order to account for some energy leaking to its neighboor frequency. If their total energy is >= than a majority percentage threshold (**80%** manually defined), we define the peak frequency as a "Whistled" moment and add to the list.  

<img src="windowed.png" style="width=10px">
  
At the end, we only select whistled moments because that's what we care about. If silence was a criteria for the whistling comparison, we'd have to find another way to extract key moments from the spectrogram.  
  
To give a final fancy touch, we smooth even more the whistled frequencies by applying a moving average to the final whistled frequencies. We expect that this move at end produce **even smoother synthesised whistles**.  

```python
# get peak from each window and form an estimated melody. I NEED TO ADD FREQUENCIES ONLY WHEN THERE IS A WHISTLE!

def moving_average(a, n=3) :
        ret = np.cumsum(a, dtype=float)
        ret[n:] = ret[n:] - ret[:-n]
        return ret[n - 1:] / n
    
def peak_extraction(Zxx, f, print_peak=False):
    Zxx_analysis = np.transpose(Zxx)
    threshold_energy = .8 # whistle section must have 80% of the whole fourier window
    peak_width = 3
    top_freqs = []
    for i in range(len(Zxx_analysis)):
        magnitudes = np.square(Zxx_analysis[i])
        highest_magnitude_index = np.argmax(magnitudes)
        mag_with_neighboors = np.sum(magnitudes[highest_magnitude_index - peak_width: highest_magnitude_index + peak_width])

        if(mag_with_neighboors > threshold_energy * np.sum(magnitudes)):
            top_freqs.append(f[highest_magnitude_index])

    top_freqs = np.array(top_freqs)
    top_freqs = moving_average(top_freqs,5)

    if(print_peak == True):
        plt.plot(top_freqs, '*-');
    return top_freqs

top_freqs2 = peak_extraction(Zxx1, f1, True)
    
```

## Chromagrams

Identification of pitches in the signal and use music templates to identify the possible keys. In the case below, the identified notes were similar to the proposed - `[D6, G6, D6, F6]` - without duplicates. The algorithm below detected `['D#6', 'G6', 'D#6', 'F#6']`. This happens either because of how humans don't control their whistle so well or because of calibration errors in the pitch detection algorithm.

```python
def get_notes_frequency(signal_peaks, n_half_steps_threshold=2):
    # TODO: evaluate for frequency stability
    
    HALF_STEP = 2**(1/12)      # half step %
    THRESHOLD_N_HALF_STEPS = n_half_steps_threshold # 
    THRESHOLD_FREQ = HALF_STEP ** THRESHOLD_N_HALF_STEPS

    notes = []
    seq = [signal_peaks[0]]

    for s in signal_peaks[1:]:
        # if there is a higher jump than half step from the average past notes, add previous note to final list
        # and restart process
        if (s > (np.mean(seq) * THRESHOLD_FREQ)) | (s < (np.mean(seq) / THRESHOLD_FREQ)):
            notes.append(np.mean(seq))
            seq = [s]
        else:        
            seq.append(s)
    # add last notes section
    notes.append(np.mean(seq))
    return notes


##########################
# https://www.johndcook.com/blog/2016/02/10/musical-pitch-notation/

A4 = 440
C0 = A4*pow(2, -4.75)
name = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]
    
def get_notes_pitch(freq):
    freq = np.array(freq)
    h = np.round(12*np.log2(freq/C0))
    octave = np.floor_divide(h, 12).astype(int)
    n = (h % 12).astype(int)
    return [name[n_] + str(octave_) for n_,octave_ in zip(n,octave)]
        
notes_freq = get_notes_frequency(top_freqs2)
print(f'frequencies {notes_freq}')

notes_pitch = get_notes_pitch(notes_freq)
print(f'pitches {notes_pitch}')
```

# Normalisation by speed


Dynamic Time Warping is a method to align two time series, providing the best alignment measure between them. As the matching is not linear, DTW can account for signals in different speed and even on different lengths, turning as an optimal reference when we're calculating similarities in any sequence data, from videos to sound. Be careful that this isn't a **distance** measure, because it doesn't guarantee that dist(A,B) = dist(B,A) (triangle inequality).  
  
DTW algorithm works with dynamic programming concept, *i.e.*, it tries to break down a complex problem by dividing it into small subproblems and recording the results of each part, so it can access them later to join and compute the final operation at the end. There's a great explanation of it's inner working on this [youtube video](https://www.youtube.com/watch?v=_K1OsqCicBY) and if you're not sure of how it works, I encourage you to check it out :)
  
  
In order to choose what is a suitable distance, we based ourselves on the [original notebook](https://github.com/prandoni/COM303-Py3/blob/master/WhistlingKeylock/WhistlingKeylock.ipynb), where the author used a combination of variations of the same melody compared to different sounds and estimated the minimum distance metric to say whether two whistles were the same. Probably because we used a out of the box DTW algorith, the distances changed a bit, but the idea remains the same.
  
Lets calculate the distance between our 2 whistles for an example and see their alignment:

```python
distance, path = fastdtw(top_freqs2_n, top_freqs3_n, dist=euclidean)

m = np.zeros([len(top_freqs2_n), len(top_freqs3_n)])

for pair in path:
    m[pair[0], pair[1]] = 1
    
plt.imshow(m, origin='lower', aspect='auto')
plt.xlabel('Whistle 1')
plt.ylabel('Whistle 2')
plt.title(f'DTW Alignment for Whistle 1 and 2. Distance: {np.round(distance,2)}');
```

# Practice yourself!

With all pieces on place, we can put everything together and create a function that, given an original audio file, records a 3 seconds whistle, compare them and produce a final similarity score by executing all the steps we've seen above.

```python
THRESHOLD = 8

def compare_with_original(sec, Fs, original_audio):
    
    f0, t0, Zxx0 = stft_music(original_audio, Fs)
    top_freqs0 = peak_extraction(Zxx0,f0,False)
    norm_freq0 = top_freqs0[int(len(top_freqs0) * .3)] # get the stable frequency at 1/3 of the frequency peaks
    top_freqs0 = top_freqs0/norm_freq0
    print(get_notes_pitch(get_notes_frequency(top_freqs0)))
    
    in_key = sd.rec(int(sec * Fs), samplerate=Fs, channels=1, blocking=True).flatten()
    f, t, Zxx = stft_music(in_key, Fs)
    top_freqs = peak_extraction(Zxx,f,False)
    norm_freq = top_freqs[int(len(top_freqs) * .3)]
    top_freqs = top_freqs/norm_freq
    print(get_notes_pitch(get_notes_frequency(top_freqs)))
    
    sim, path = fastdtw(top_freqs0, top_freqs, dist=euclidean)

    plt.plot(top_freqs0, '*-')
    plt.plot(top_freqs, '*-')
    plt.legend(['Original', 'Recorded']);
    
    print(f'Similarity - {sim}')
    if(sim <= THRESHOLD):
        print('OPEN SESAME!')
    else:
        print('No lunch for you :(')
```
