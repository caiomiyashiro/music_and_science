```python
import numpy as np
import matplotlib.pyplot as plt

from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 15, 4
```

# Discrete Fourier Transform (DFT) Formula

$$X[k] = \sum_{n = 0}^{N - 1}x[n]e^{-j2\pi kn/N}, \space \space k = 0, 1,..., N-1$$

The discrete fourier transform get a signal of length N and transform into another signal of same length. However, each index k in the transform shows the magnitude (presence) of a senoid of frequency $\frac{F_sk}{N}$ Hz, where $F_s$ is the sampling frequency.  
  
So, depending on the signal length and sample frequency, each index k will represent a different frequency presence.  
  
To understand the fourier transform, we can look at the weird part first, the complex exponential:  

# Complex Exponentials are just senoids

Each point in the complex exponential is just a point but see that we are multipliying each point from the signal by a complex exponential signal with the same length as the original signal, basically a dot product.  
  
If we take a look at all the points that form the complex exponential, we see that it's basically a senoid. Of course, we are simplifying this first section because we will be ignoring the imaginary part. But we'll come back to it in the future.

```python
index = np.arange(10000)
FREQ = 5


Fs = len(index)
N = len(index)
k = (FREQ/Fs) * N

print(f'Freq Hz = {Fs*k/N}')
plt.plot(np.exp(-1j * 2 * np.pi * k * index / N))
plt.title('Complex Exponential for k = 0 to N-1');
plt.ylabel('Amplitude');
```

# DFT is just a series of dot products
  
By seeing the complex exponential as just a senoid, we can abstract the formula to just a single dot product between the original signal and the senoid for each frequency index k, from k = 0 to N-1:

$$X[k] = <signal, complex\_exponential(k)>, \space \space k = 0, 1,..., N-1$$  
  
By performing the dot product, dot products will be bigger for frequencies that the signal respond more. For instance, consider a senoid made of frequency 20 and 70 Hz and amplitudes equal to 4 and 7 respectively:  

```python
Fs = 10000
n = np.arange(Fs*3)

T = 1/Fs

# senoid 1
f = 20
signal_1 = np.cos(2 * np.pi * f * n * T) * 4

# senoid 2
f = 70
signal_2 = np.cos(2 * np.pi * f * n * T) * 7 

signal = signal_1 + signal_2

plt.plot(signal);
```

First, lets take a look at the final response, the DTF result:  

https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.fft.fft.html#numpy.fft.fft

```python
from numpy.fft import fft

half_signal = signal[:int(len(signal)/2)]

freqs= Fs*np.arange(len(signal))/len(signal)
freqs = freqs[:int(len(half_signal))]    # first half of the frequencies

signal_ = (signal)/(len(half_signal))    # normalisation factor

zoom = 300
plt.plot(freqs[:zoom], abs(fft(signal_)[:zoom]))
plt.title('Frequencies from Senoid = 20 + 70 HZ')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude');
```

Because the signal is only a mixture of two pure senoids, wee see that it responds highly to these specific frequencies.

So a signal that varies in the same frequency as the senoids will have a bigger magnitude in the final transformation.
